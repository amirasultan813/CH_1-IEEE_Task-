[Problem 1](https://leetcode.com/problems/remove-all-adjacent-duplicates-in-string/)

```python
class Solution(object):
    def removeDuplicates(self, s):
        if not s:
            return s
        
        res = [s[0]]
        for i in s[1:]:
            if res and i == res[-1]:
                res.pop()
            else:
                res.append(i)
                
        return "".join(res)
```
